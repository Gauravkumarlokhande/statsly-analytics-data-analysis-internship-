# Multivariate Plots

```python
def imputer(value):
    if value=='subject1'or value=='subject2' or value=='subject3' or value=='subject4' or value=='subject5':
        return 'subject groupb 1'
    else:
        return 'subject group 2'
```

```python
data['subject']=data['subject'].apply(imputer)
```

```python
data['subject'].unique()
```

```python
def impute(value):
    if value<=6:
        return 'activity group 1'
    else:
        return 'activity group 2'
```

```python
data['Activity']=data['Activity'].apply(impute)
```

```python
data['Activity'].unique()
```
