```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Red Wine Quality.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['quality'].unique()
```

```python
def quality(value):
    if value>=3 and value<=5:
        return 'bad'
    else:
        return 'good'
```

```python
data['quality']=data['quality'].apply(quality)
```

```python
data['quality'].value_counts()
```

# Distribution plots for all variables

```python
for i in data.drop(columns=['quality']).columns:
    plt.figure(figsize=(10,8))
    sns.distplot(data[i])
```

# Boxplots

```python
for i in data.drop(columns=['quality']).columns:
    plt.figure(figsize=(10,8))
    sns.boxplot(data[i],color='orange')
```

# Remove Outliers

```python
def impute(value):
    if value<lower:
        return lower
    elif value>upper:
        return upper
    else:
        return value
```

```python
for i in data.drop(columns=['quality']).columns:
    df=data[i].describe()
    q1=df[4]
    q3=df[6]
    iqr=q3-q1
    lower=q1-1.5*iqr
    upper=q3+1.5*iqr
    data[i]=data[i].apply(impute)
    
```

```python
for i in data.drop(columns=['quality']).columns:
    plt.figure(figsize=(10,8))
    sns.boxplot(data[i],color='blue')
```

# Bivariate Plots

```python
for i in data.drop(columns=['quality']).columns:
    plt.figure(figsize=(10,8))
    data.groupby('quality')[i].mean().plot(kind='bar',color='green')
    plt.ylabel(i)
    plt.title('quality of the wine and {}'.format(i))
```

# Training and testing the model 

```python
from sklearn.model_selection import train_test_split as tst
```

```python
x=data.drop(columns=['quality'])
y=data['quality']
```

```python
x_train,x_test,y_train,y_test=tst(x,y,test_size=0.3,random_state=101)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler as SC
```

```python
std=SC()
```

```python
x_train=std.fit_transform(x_train)
x_test=std.fit_transform(x_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
print("Training accuracy:",lr.score(x_train,y_train))
print("Testing accuracy:",lr.score(x_test,y_test))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier as DTC
```

```python
dt=DTC()
```

```python
dt.fit(x_train,y_train)
```

```python
print('Training score:',dt.score(x_train,y_train))
print('Testing score:',dt.score(x_test,y_test))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier as RFC 
```

```python
rf=RFC()
```

```python
rf.fit(x_train,y_train)
```

```python
print('Training score:',rf.score(x_train,y_train))
print('Testing score:',rf.score(x_test,y_test))
```

# SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
sv=SVC()
```

```python
sv.fit(x_train,y_train)
```

```python
print('Training score:',sv.score(x_train,y_train))
print('Testing score:',sv.score(x_test,y_test))
```

```python
y_pred_train=sv.predict(x_train)
y_pred_test=sv.predict(x_test)
```

```python
cr=classification_report(y_pred_train,y_train)
cr1=classification_report(y_pred_test,y_test)
```

```python
print(cr)
print(cr1)
```

```python
cm=confusion_matrix(y_pred_test,y_test)
cm
```
